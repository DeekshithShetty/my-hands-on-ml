## Matplotlib example workflow

```python
 # A matplotlib workflow

# 0. Import and get matplotlib ready
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10,10))

# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Sample Simple Plot", xlabel="x-axis", ylabel="y-axis")

# 5. Save & show
fig.savefig("./images/simple-plot.png")
```

## Making figures with NumPy arrays

We want:
* Line plot (default)
* scatter
* bar
* hist
* subplots()

```python
import numpy as np

# Create some data
x = np.linspace(0, 10, 100)
x[:10]
```

```python
# Plot the data and create a line plt
fig, ax = plt.subplots()
ax.plot(x, x**2)
```

```python
# scatter plt
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
my_dict = {
    "a": 10,
    "b": 2,
    "c": 18
}
fig, ax = plt.subplots()
ax.bar(my_dict.keys(), my_dict.values());
```

```python
x = np.random.randn(1000) # std normal distribution
fig, ax = plt.subplots()
ax.hist(x);
```

## Two options for subplots

```python
# subplot option 1 (we use this)
fig, ((ax0, ax1), (ax2, ax3)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10, 5))

ax0.plot(x, x/2)
ax1.scatter(np.random.random(10), np.random.random(10))
ax2.bar(my_dict.keys(), my_dict.values());
ax3.hist(np.random.randn(1000));
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(10, 5))

ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].bar(my_dict.keys(), my_dict.values());
ax[1, 1].hist(np.random.randn(1000));
```

## Plotting from pandas DataFrames

```python
import pandas as pd
```

```python
car_sales = pd.read_csv("./data/car-sales.csv")
car_sales
```

```python
# Start with some dummy data
ts = pd.Series(np.random.randn(1000),
               index=pd.date_range('1/1/2020', periods=1000))
ts = ts.cumsum()
ts.plot()
```

```python
# Remove price column symbols
car_sales["Price"] = car_sales["Price"].str.replace('[\$\,\.]', '')
car_sales
```

```python
# Remove last two zeros
car_sales["Price"] = car_sales["Price"].str[:-2]
car_sales
```

```python
# Add a date column
car_sales["Sale Date"] = pd.date_range("1/1/2020", periods=len(car_sales))
car_sales
```

```python
car_sales["Total Sales"] = car_sales["Price"].astype(int).cumsum()
car_sales
```

```python
car_sales.plot(x="Sale Date", y="Total Sales")
```

```python
# Convert Price to int
car_sales["Price"] = car_sales["Price"].astype(int)
car_sales.plot(x="Odometer (KM)", y="Price", kind='bar');
```

```python
 car_sales["Price"].plot.hist(bins=10);
```

```python
heart_disease = pd.read_csv("./data/heart-disease.csv")
heart_disease.head()
```

```python
# Perform data analysis on patients over 50
over_50 = heart_disease[heart_disease["age"] > 50]
over_50
```

```python
over_50.plot(kind='scatter', 
             x='age', 
             y='chol', 
             c='target', 
             figsize=(10, 6));
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind='scatter', 
             x="age", 
             y="chol", 
             c='target', 
             ax=ax);
ax.set_xlim([45, 100]);
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
scatter = ax.scatter(over_50["age"], 
                     over_50["chol"], 
                     c=over_50["target"])
ax.set_xlim([45, 100]);
```

```python
# Make a bit more of a complicated plot

# Create the plot
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(over_50["age"], 
                     over_50["chol"], 
                     c=over_50["target"])

# Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol");
ax.legend(*scatter.legend_elements(), title="Target");

# Add a meanline
ax.axhline(over_50["chol"].mean(),
           linestyle="--");
```

## Adding another plot to existing styled one

```python
# Setup plot (2 rows, 1 column)
fig, (ax0, ax1) = plt.subplots(nrows=2, # 2 rows
                               ncols=1, 
                               sharex=True, 
                               figsize=(10, 8))

# Add data for ax0
scatter = ax0.scatter(over_50["age"], 
                      over_50["chol"], 
                      c=over_50["target"])
# Customize ax0
ax0.set(title="Heart Disease and Cholesterol Levels",
        ylabel="Cholesterol")
ax0.legend(*scatter.legend_elements(), title="Target")

# Setup a mean line
ax0.axhline(y=over_50["chol"].mean(), 
            color='b', 
            linestyle='--', 
            label="Average")


# Add data for ax1
scatter = ax1.scatter(over_50["age"], 
                      over_50["thalach"], 
                      c=over_50["target"])

# Customize ax1
ax1.set(title="Heart Disease and Max Heart Rate Levels",
        xlabel="Age",
        ylabel="Max Heart Rate")
ax1.legend(*scatter.legend_elements(), title="Target")

# Setup a mean line
ax1.axhline(y=over_50["thalach"].mean(), 
            color='b', 
            linestyle='--', 
            label="Average")

# Title the figure
fig.suptitle('Heart Disease Analysis', fontsize=16, fontweight='bold');
```
