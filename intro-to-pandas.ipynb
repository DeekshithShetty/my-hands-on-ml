```python
# shuffle the data
shuffled_car_sales = car_sales.sample(frac=1)
shuffled_car_sales

# however this function is a slow for millions of data in a dataset
# best way is to fraction small amount of data and do experiments
# eg car_sales.sample(frac=0.01)
```
